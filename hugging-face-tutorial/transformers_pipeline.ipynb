```python
import transformers
```

**Pipelining in Hugging Face**

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
# alternatively
# classifier = pipeline("text-classification")
classifier("The Lord of the Rings is an absolute classic.")
```

*Zero-shot classification pipeline*

```python
zero_shot_classifier = pipeline("zero-shot-classification")

print(zero_shot_classifier("The Monalisa is masterful creation",
                           candidate_labels=["movies", "art", "safety"]))
```

*Text-generation*

```python
generator = pipeline("text-generation")
print(generator("JRR Tolkien was a great linguist and"))
```

```python
generator = pipeline("text-generation", model="distilgpt2")
generator("JRR Tolkien was a great linguist and", max_length=30, num_return_sequences=2)
```

*Mask-filling*

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

*Question-Answering*

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```
